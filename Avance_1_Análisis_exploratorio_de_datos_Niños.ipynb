```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargando el archivo Excel proporcionado para el análisis
file_path_ninos_eda = '/content/2.2.Base de ACES_Menores.xlsx'
df_ninos_eda_completo = pd.read_excel(file_path_ninos_eda)

# Columnas seleccionadas para el análisis
columnas_seleccionadas = [
    "ENTIDAD", "MUN", "MUNICIPIO", "LOC", "LOCALIDAD", "NN_M04_05", "NN_M04_09", "NN_M04_11",
    "NN_M04_12", "NN_M04_13", "NN_M08_15", "NN_M08_16", "NN_M08_17", "NN_M08_17A", "NOTA9",
    "NN_M09_01", "NN_M09_02", "NN_M09_03", "NN_M09_04", "NOTA10", "NN_M10_01", "NN_M10_01A",
    "LIBROS", "NN_M10_02", "NN_M10_03", "DIAS_SOLO", "NN_M10_04A", "NN_M10_04B", "NN_M10_04C",
    "NN_M10_04D", "NN_M10_04E", "NN_M10_04F", "NN_M10_04G", "NN_M10_05A", "NN_M10_05B",
    "NN_M10_05C", "NN_M10_05D", "NN_M10_05E", "NN_M10_05F", "NN_M10_05G", "NN_M10_05H",
    "NN_M10_05I", "NN_M10_05J", "NN_M10_05K", "DISCIPLINA", "NN_M10_06", "PEARLS1", "Escala1",
    "Escala2", "Escala3", "Escala4", "Escala5", "Escala_Inter", "Escala_Exter", "Escala_Total",
    "SDQ3", "DxSDQ3", "SDQ4", "DxSDQ4", "PEARLS_Pte2", "ABUSO_SEXUAL", "AREA", "EDADM",
    "GPOEDAD", "ESCOLARIDAD", "HERMANOS", "ACT_FAM", "NLIBROS", "NOVIOLENCIA", "VIOLENCIA_FIS",
    "VIOLENCIA_PSIC", "VIOLENCIA", "VIOLENCIA_SEVERA"
]

# Cargar el conjunto de datos con las columnas seleccionadas
df_ninos_eda = pd.read_excel(file_path_ninos_eda, usecols=columnas_seleccionadas)

# Vista previa de los datos
df_ninos_eda.head()
```

```python
# Identificación de valores faltantes en las columnas seleccionadas
valores_faltantes_ninos = df_ninos_eda_completo[columnas_seleccionadas].isnull().sum()

# Estadísticas resumidas del conjunto de datos
estadisticas_resumidas_ninos = df_ninos_eda_completo[columnas_seleccionadas].describe(include='all')

# Visualización de valores faltantes usando un mapa de calor
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
sns.heatmap(df_ninos_eda_completo[columnas_seleccionadas].isnull(), cbar=False)
plt.title('Mapa de Calor de Valores Faltantes en el Conjunto de Datos de Niños')
plt.show()

valores_faltantes_ninos, estadisticas_resumidas_ninos
```

**Identificación de Valores Faltantes:**

Hay valores faltantes en varias columnas. Por ejemplo, las columnas "NN_M04_12" y "GPOEDAD" tienen un número significativo de valores faltantes.
El mapa de calor muestra la distribución de los valores faltantes en todo el conjunto de datos, lo que ayuda a visualizar en qué columnas faltan más datos.
Estadísticas Resumidas:

Las estadísticas resumidas proporcionan una visión general de las características del conjunto de datos, como el rango, la media y la desviación estándar de las variables numéricas, así como la frecuencia de las categorías en las variables categóricas.

```python
df_ninos_eda_especifico = df_ninos_eda_completo[columnas_seleccionadas]

# Análisis de Valores Atípicos
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_ninos_eda_especifico.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.title('Análisis de Valores Atípicos para Variables Numéricas Seleccionadas')
plt.show()

# Cardinalidad de Variables Categóricas
cardinalidad_categoricas = df_ninos_eda_especifico.select_dtypes(include=['object']).nunique()
cardinalidad_categoricas


```

```python
# Distribuciones y Posibles Transformaciones
# Ejemplo con una variable numérica
plt.figure(figsize=(8, 5))
sns.histplot(df_ninos_eda_especifico['SDQ3'].dropna(), kde=True)
plt.title('Distribución de la Variable SDQ3')
plt.show()

# Correlaciones entre Variables Numéricas
correlaciones = df_ninos_eda_especifico.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlaciones, annot=False, cmap='coolwarm')
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```

```python

#Limpieza de datos
# Eliminar columnas con un alto porcentaje de valores faltantes

threshold = 0.5 # Porcentaje de valores faltantes para eliminar una columna
columns_to_drop = valores_faltantes_ninos[valores_faltantes_ninos > threshold * len(df_ninos_eda)].index
data_cleaned = df_ninos_eda.drop(columns=columns_to_drop)

# Eliminar filas con cualquier valor faltante
data_cleaned=data_cleaned.dropna()

categorical_cols = data_cleaned.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    # One-hot encoding si la cardinalidad es baja (menos de 10 categorías únicas)
    if data_cleaned[col].nunique() < 10:
        data_cleaned = pd.get_dummies(data_cleaned, columns=[col], drop_first=True)
    else:
        # Para simplificar, en este caso solo eliminaremos las columnas con alta cardinalidad
        data_cleaned = data_cleaned.drop(columns=[col])

# Verificar los cambios realizados
data_cleaned.info()
```

**Valores Faltantes:** Se eliminaron las columnas con más del 50% de valores faltantes, y luego se eliminaron las filas que todavía tenían valores faltantes. Esto resultó en un conjunto de datos más limpio pero también más pequeño.

```python
# Estadísticas Descriptivas
descriptive_stats_cleaned = data_cleaned.describe()

# Detección de Valores Atípicos en el conjunto de datos limpio
# Calcular el rango intercuartílico (IQR) para detectar valores atípicos
Q1_cleaned = descriptive_stats_cleaned.transpose()['25%']
Q3_cleaned = descriptive_stats_cleaned.transpose()['75%']
IQR_cleaned = Q3_cleaned - Q1_cleaned
outliers_cleaned = ((data_cleaned < (Q1_cleaned - 1.5 * IQR_cleaned)) | (data_cleaned > (Q3_cleaned + 1.5 * IQR_cleaned))).sum()

# Resumen de los resultados obtenidos para valores atípicos
outliers_summary_cleaned = outliers_cleaned[outliers_cleaned > 0].sort_values(ascending=False)

(descriptive_stats_cleaned.transpose().head(10), outliers_summary_cleaned.head(10))
```

```python
# Crear una matriz de correlación para las variables numéricas
correlation_matrix = data_cleaned.corr()

# Configuración de visualización
plt.figure(figsize=(12, 10))
sns.set(font_scale=0.5)
# Crear un mapa de calor para visualizar la matriz de correlación
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, linewidths=.5)
plt.title('Matriz de Correlación de Variables Numéricas')

# Mostrar el gráfico
plt.show()
```
